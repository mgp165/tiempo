# Metrics reporting

About linear trees

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading files

tmax_lt = pd.read_parquet('lt_tmax.parquet')
tmin_lt = pd.read_parquet('lt_tmin.parquet')
prec_lt = pd.read_parquet('lt_prec.parquet')
sol_lt = pd.read_parquet('lt_sol.parquet')
```

## MAE

```python
fig, axs = plt.subplots(1, 4, figsize=(16, 4))

axs[0].plot(tmax_lt.Train_MAE, label='Train')
axs[0].plot(tmax_lt.Test_MAE, label='Test')
axs[0].legend()
axs[0].set_title('MAE tmax')

axs[1].plot(tmin_lt.Train_MAE, label='Train')
axs[1].plot(tmin_lt.Test_MAE, label='Test')
axs[1].legend()
axs[1].set_title('MAE tmin')

axs[2].plot(prec_lt.Train_MAE, label='Train')
axs[2].plot(prec_lt.Test_MAE, label='Test')
axs[2].legend()
axs[2].set_title('MAE prec')

axs[3].plot(sol_lt.Train_MAE, label='Train')
axs[3].plot(sol_lt.Test_MAE, label='Test')
axs[3].legend()
axs[3].set_title('MAE sol')

plt.show()
```

### tmax: 
    No cambia nada a medida que aumenta la profundidad.
    De media se comete error de 1.5 grados en Train y casi 2 en Test. (error asumible)

### tmin:
    No cambia nada a medida que aumenta la profundidad.
    De media se comete error de 1 grado aproximadamente en Train y 1.5 en Test. (error asumible)
    
### prec:
    El que mejor va es el de profundidad 5.
    En Train de media se comete error de 1.5 mm y en Test, casi 2mm. (error asumible)

### sol:
    No cambia nada a medida que aumenta la profundidad.
    De media se comete error de 2 horas aproximadamente en Train y 2.5 en Test. (error grande)

Obviamente a más profundidad más tiempo de ejecución.

## Delta

```python
fig, axs = plt.subplots(1, 4, figsize=(16, 4))

axs[0].plot(tmax_lt.delta)
axs[0].set_title('Delta tmax')

axs[1].plot(tmin_lt.delta)
axs[1].set_title('Delta tmin')

axs[2].plot(prec_lt.delta)
axs[2].set_title('Delta prec')

axs[3].plot(sol_lt.delta)
axs[3].set_title('Delta sol')

plt.show()
```

### tmax: 
    No cambia nada a medida que aumenta la profundidad.
    De media se comete error de 0.5 grados aproximadamente más en Test que en Train. (asumible)

### tmin:
    No cambia nada a medida que aumenta la profundidad.
    De media se comete error de 0.25 grados aproximadamente más en Test que en Train.(asumible)
    
### prec:
    El de profundidad 5 generaliza un poco mejor.
    De media cometemos error entre 0.3 y 0.4 mm más en Test que en Train. (asumible)

### sol:
    No cambia nada a medida que aumenta la profundidad.
    De media se comete error de 0.37 horas aproximadamente más en Test que en Train. (asumible)

## Run time

```python
fig, axs = plt.subplots(1, 4, figsize=(16, 4))

axs[0].plot(tmax_lt.Run_Time)
axs[0].set_title('Run Time tmax')

axs[1].plot(tmin_lt.Run_Time)
axs[1].set_title('Run Time tmin')

axs[2].plot(prec_lt.Run_Time)
axs[2].set_title('Run Time prec')

axs[3].plot(sol_lt.Run_Time)
axs[3].set_title('Run Time sol')

plt.show()
```

### tmax
    El más rápido es el de profundidad 5 (1min) y el más lento el de 15 (1min 14s).
### tmin
    El más rápido es el de profundidad 5 (1min 8s) y el más lento el de 15 (1min 13s).
### prec
    El más rápido es el de profundidad 5 (1min 24s) y el más lento el de 10 (1min 27s).
### sol
    El más rápido es el de profundidad 15 (1min) y el más lento el de 10 (1min 6s).
