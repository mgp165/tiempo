```python
import pandas as pd
```

```python
# lista de archivos JSON
archivos_json = ['1970-1975_alicante_alicante.json', '1975-1980_alicante_alicante.json', '1980-1985_alicante_alicante.json',
                             '1985-1990_alicante_alicante.json', '1990-1995_alicante_alicante.json', '1995-2000_alicante_alicante.json',
                             '2000-2005_alicante_alicante.json', '2005-2010_alicante_alicante.json', '2010-2015_alicante_alicante.json', 
                             '2015-2020_alicante_alicante.json', '2020-2023_alicante_alicante.json']
```

```python
# lista para almacenar los DataFrames de 5 a√±os
dataframes = []

# leer y convertir los archivos JSON en DataFrames
for archivo in archivos_json:
    df = pd.read_json(archivo)
    dataframes.append(df)

# Combinar los DataFrames en el orden deseado
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
# formato de fecha 
combined_df['fecha'] = pd.to_datetime(combined_df.fecha, format='%Y-%m-%d')
combined_df['horatmin'] = pd.to_datetime(combined_df.horatmin, format='%H:%M', errors='coerce').dt.time
combined_df['horatmax'] = pd.to_datetime(combined_df.horatmax, format='%H:%M', errors='coerce').dt.time
combined_df['horaracha'] = pd.to_datetime(combined_df.horaracha, format='%H:%M', errors='coerce').dt.time
combined_df['horaPresMax'] = pd.to_datetime(combined_df.horaPresMax, format='%H', errors='coerce').dt.time
combined_df['horaPresMin'] = pd.to_datetime(combined_df.horaPresMin, format='%H', errors='coerce').dt.time
```

```python
combined_df = combined_df.replace(',', '.', regex=True)
```

```python
combined_df.prec = combined_df.prec.replace('Ip', '0.0', regex=True)
```
