```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading files

tmax_rerf = pd.read_parquet('rerf_tmax.parquet')
tmin_rerf = pd.read_parquet('rerf_tmin.parquet')
prec_rerf = pd.read_parquet('rerf_prec.parquet')
sol_rerf = pd.read_parquet('rerf_sol.parquet')
```

```python
tmax_rerf = tmax_rerf.reset_index()
tmin_rerf = tmin_rerf.reset_index()
prec_rerf = prec_rerf.reset_index()
sol_rerf = sol_rerf.reset_index()
```

```python
tmax_rerf.head()
```

```python
tmax_rerf.alpha.unique()
```

```python
plt.figure(figsize=(20,16))

plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 0.25]['Train_MAE'], label='alpha=0.25')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 0.5]['Train_MAE'], label='alpha=0.5')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 0.75]['Train_MAE'], label='alpha=0.75')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 1]['Train_MAE'], label='alpha=1')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 1.25]['Train_MAE'], label='alpha=1.25')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 1.5]['Train_MAE'], label='alpha=1.5')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 2]['Train_MAE'], label='alpha=2')

plt.title('Train MAE')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,16))

plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 0.25]['Test_MAE'], label='alpha=0.25')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 0.5]['Test_MAE'], label='alpha=0.5')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 0.75]['Test_MAE'], label='alpha=0.75')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 1]['Test_MAE'], label='alpha=1')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 1.25]['Test_MAE'], label='alpha=1.25')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 1.5]['Test_MAE'], label='alpha=1.5')
plt.plot(list(range(84)), tmax_rerf[tmax_rerf['alpha'] == 2]['Test_MAE'], label='alpha=2')

plt.title('Test MAE')
plt.legend()
plt.show()
```

En el train se comportan todos los $\alpha$ prcticamente igual.\
En el test se ven ligeras diferencias, a mayor $\alpha$ menor MAE.
