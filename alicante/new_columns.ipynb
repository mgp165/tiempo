```python
import pandas as pd
import numpy as np
```

```python
tiempo = pd.read_parquet('dataframe_1970-2023_alicante_alicante.parquet')
```

```python
TIEMPO = tiempo
```

```python
TIEMPO
```

```python
variables_shift = ['tmed', 'prec', 'tmin', 'horatmin', 'tmax', 'horatmax', 'dir', 'velmedia', 'racha', 'horaracha', 'sol', 'presMax', 'horaPresMax', 'presMin', 'horaPresMin']
```

```python
for var in variables_shift:
    TIEMPO[var+'_1'] = TIEMPO[var].shift(1)
    TIEMPO[var+'_2'] = TIEMPO[var].shift(2)
    TIEMPO[var+'_3'] = TIEMPO[var].shift(3)
    TIEMPO[var+'_mañana'] = TIEMPO[var].shift(-1)
```

```python
TIEMPO
```

```python
TIEMPO.to_parquet('dataframe_final_1970-2023_alicante_alicante.parquet')
```
